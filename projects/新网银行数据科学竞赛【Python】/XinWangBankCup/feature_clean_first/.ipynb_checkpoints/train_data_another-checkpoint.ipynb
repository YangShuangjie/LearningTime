```python
def impute_outliers(col):
    desc = numeric_var[col].describe()
    upper = desc["75%"]+1.5*(desc["75%"]-desc["25%"])
    lower = desc["25%"]-1.5*(desc["75%"]-desc["25%"])
    outliers = numeric_var[col][(numeric_var[col]>upper)|(numeric_var[col]<lower)]
    need_values = [desc["mean"],desc["50%"],desc['25%'],desc['75%'],desc["min"],desc["max"]]
    impute_values = [need_values[i] for i in [np.random.randint(6) for _ in range(len(outliers))]]
    numeric_var[col] = pd.Series(impute_values,index=outliers.index).combine_first(numeric_var[col])

for n in range(10):
    for col in continous_var:
        impute_outliers(col)
```

# final step:feature scale and one-hot code
## feature scale

```python
def continous_var_scale(cols,method="mean_std"):
    if method == "min_max":
        train_xy[cols] = train_xy[cols].apply(lambda x:(x-x.mean())/x.std())
    elif method == "mean_std":
        train_xy[cols] = train_xy[cols].apply(lambda x:(x-x.min())/(x.max()-x.min()))
    return train_xy
```

```python
train_x = pd.read_csv('./another data/train_x.csv',index_col=["cust_group",'cust_id'])
test_all = pd.read_csv('./another data/test_all.csv',index_col=["cust_group",'cust_id'])
```

```python
from sklearn.preprocessing import OneHotEncoder

def one_hot_col(col,train_xy=train_xy,train_x=train_x,test_all=test_all):
    one_hot  = OneHotEncoder()
    one_hot.fit(train_xy[col].fillna(99).values.reshape(-1,1))
    
    xy_hot = one_hot.transform(train_xy[col].fillna(99).values.reshape(-1,1)).toarray()
    x_hot = one_hot.transform(train_x[col].fillna(99).values.reshape(-1,1)).toarray()
    test_hot = one_hot.transform(test_all[col].fillna(99).values.reshape(-1,1)).toarray()
    
    train_xy_col_hot = pd.DataFrame(xy_hot,columns=[col+"_%s" % i for i in range(xy_hot.shape[1])],
                                    index=train_xy[col].index)

    
    train_x_col_hot = pd.DataFrame(x_hot,columns=[col+"_%s" % i for i in range(x_hot.shape[1])],
                                   index=train_x[col].index)

    
    test_all_col_hot = pd.DataFrame(test_hot,columns=[col+"_%s" % i for i in range(test_hot.shape[1])],
                                    index=test_all[col].index)

#     col_names_xy = [col+'_%s' % i for i in range(int(train_xy[col].min()),int(train_xy[col].max()+1))]+['nan']
#     col_names_x = [col+'_%s' % i for i in range(int(train_x[col].min()),int(train_x[col].max()+1))]+['nan']
#     col_names_ = [col+'_%s' % i for i in range(int(test_all[col].min()),int(test_all[col].max()+1))]+['nan']

    return train_xy_col_hot,train_x_col_hot,test_all_col_hot

def get_dummies(df):
    dummy_cols = df.loc[:,"x_96":'x_157'].columns.drop(['x_141','x_147','x_153','x_155'])
    one_hot_df = pd.get_dummies(df[dummy_cols].astype("object"),dummy_na=True)
    df = pd.concat([df.loc[:,:'x_95'],one_hot_df],axis=1)
    df[["x_1","x_2","x_93","x_95"]] = df[["x_1","x_2","x_93","x_95"]].apply(lambda x: x.fillna(x.mean()))
    df.fillna(99,inplace=True)
    return df
```

```python
for col in ['x_141','x_147','x_153','x_155']:
    train_xy_col_hot,train_x_col_hot,test_all_col_hot = one_hot_col(col)
    train_xy = pd.concat([train_xy,train_xy_col_hot],axis=1)
    train_x = pd.concat([train_x,train_x_col_hot],axis=1)
    test_all = pd.concat([test_all,test_all_col_hot],axis=1)
```

```python
train_xy_clean = get_dummies(train_xy)
train_x_clean = get_dummies(train_x)
test_all_clean = get_dummies(test_all)
```
